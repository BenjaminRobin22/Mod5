```python
contacts = [['Mike Green', 'demo1@deitel.com', '55555555555'],
            ['Ben Robin', 'demo2deitel.com', '55555551234']]
contactsdf = pd.DataFrame(contacts,
                          columns=['Name', 'Email', 'Phone'])
contactsdf
```

```python
import re
```

```python
def get_formatted_phone(value):
    result = re.fullmatch(r'(\d{})(\d{3})(\d{4})', value)
    return '-'.join(result.groups()) if result else value
```

```python
formatted_phone = contactsdf['Phone'].map(get_formatted_phone)
formatted_phone
```

```python
contactsdf['Phone'] = formatted_phone
contactsdf
```

```python
contacts = [['Mike Green', 'Michael.Green2@deitel.com', '6365555555'],
            ['Ben Robin', 'Benjamin.Robin2@gmail.com', '8165551234']]

contactsdf = pd.DataFrame(contacts,
                          columns=['Name', 'Email', 'Phone'])

def get_formatted_phone(value):
    result = re.fullmatch(r'(\d{3})(\d{3})(\d{4})',value)
    if result:
        part1, part2, part3 = result.groups()
        return '(' +part1 + ') ' + part2 + '-' + part3
    else:
        return "error"
```

```python
contactsdf['Phone'] = contactsdf['Phone'].map(get_formatted_phone)
contactsdf
```
